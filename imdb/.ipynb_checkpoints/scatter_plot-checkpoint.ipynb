```python
from metric.retrieval_eval import *
from sklearn.manifold import TSNE
import numpy as np
%matplotlib inline
############################################################
import matplotlib
matplotlib.rcParams['ps.useafm'] = True
matplotlib.rcParams['pdf.use14corefonts'] = True
matplotlib.rcParams['text.usetex'] = True
############################################################
from matplotlib import pyplot as plt
import json
from matplotlib.backends.backend_pdf import PdfPages
from sklearn.metrics import pairwise_distances
```

```python
class IndexedMatrix():
    def __init__(self, items, repr_matrix):
        self.items = np.array(items)
        self.repr_matrix = np.array(repr_matrix).astype(np.float32)
 
 
def gen_indexed_matrix(items, repr_dict):
    repr_matrix = [repr_dict[item] for item in items]
    return IndexedMatrix(items, repr_matrix)

def id2sth(key, path):
    with open(path) as f:
        ls_dict = json.load(f)
    return {elem["movieId"]:elem[key] for elem in ls_dict}
```

```python
data_p = "OMDB_dataset/OMDB.json"
wice_item_p = "sample_sensi/wice_10x5_1/item.embd.40"
wice_word_p = "sample_sensi/wice_10x5_1/word.embd.40"
bpt_item_p = "task/retrieval/bpt_top10_1/item.embd.40"
bpt_word_p = "task/retrieval/bpt_top10_1/word.embd.40"
```

```python
id2genre = {e:gs.split(", ") for e, gs in id2sth("genre", data_p).items()}
wice_item = read_w2v_from_file(wice_item_p)
wice_word = read_w2v_from_file(wice_word_p)
bpt_item = read_w2v_from_file(bpt_item_p)
bpt_word = read_w2v_from_file(bpt_word_p)
tsne = TSNE()
```

# sole-genre

```python
genres = ["Horror", "Sci-Fi", "Action", "Thriller", "Western", "Short"]
sole_genre_dict = {}
for genre in genres:
    sole_genre_dict[genre] = ["m_"+e for e, gs in id2genre.items() if genre in gs and "m_"+e in bpt_item and len(gs)==1]
```

```python
for g in genres:
    print(g, len(sole_genre_dict[g]))
```

```python
high_dim_embd = []
for genre in genres:
    for e_id in sole_genre_dict[genre]:
        high_dim_embd.append(bpt_item[e_id])
high_dim_embd = np.array(high_dim_embd).astype(np.float32)
tsne_embd = tsne.fit_transform(high_dim_embd)

pp = PdfPages('BPT_6gen_figure_ty1font.pdf')
plt.figure(figsize=(10,8))
plt.clf()
plt.axis('off')
plt.margins(0.0)
plt_dict = {"Horror":("blue", "."), 
            "Action":("red", "x"),
            "Sci-Fi":("green", "^"),
            "Thriller":("aqua", "*"),
            "Western":("darkorange", "s"), 
            "Short":("black", "+")
            }
collection = []
left = 0
num = 100
for g in genres:
    right = left+len(sole_genre_dict[g])
    tar = tsne_embd[left:right]
    obj = plt.scatter(tar[:num, 0], tar[:num, 1], color=plt_dict[g][0], s=50, marker=plt_dict[g][1], label=g, alpha=1)
    collection.append(obj)
    left = right
    
plt.legend(handles=collection, loc=4, fontsize=20, scatterpoints=1)
pp.savefig(bbox_inches="tight")
pp.close()
plt.tight_layout()
```

```python
high_dim_embd = []
for genre in genres:
    for e_id in sole_genre_dict[genre]:
        high_dim_embd.append(wice_item[e_id])
high_dim_embd = np.array(high_dim_embd).astype(np.float32)
tsne_embd = tsne.fit_transform(high_dim_embd)

pp = PdfPages('WICE_6gen_figure_ty1font.pdf')
plt.figure(figsize=(10,8))
plt.clf()
plt.axis('off')
plt.margins(0.0)
plt_dict = {"Horror":("blue", "."), 
            "Action":("red", "x"),
            "Sci-Fi":("green", "^"),
            "Thriller":("aqua", "*"),
            "Western":("darkorange", "s"), 
            "Short":("black", "+")
            }
collection = []
left = 0
num = 100
for g in genres:
    right = left+len(sole_genre_dict[g])
    tar = tsne_embd[left:right]
    obj = plt.scatter(tar[:num, 0], tar[:num, 1], color=plt_dict[g][0], s=50, marker=plt_dict[g][1], label=g, alpha=1)
    collection.append(obj)
    left = right
    
plt.legend(handles=collection, loc=4, fontsize=20, scatterpoints=1)
pp.savefig(bbox_inches="tight")
pp.close()
plt.tight_layout()
```

# Godzilla

```python
id2title = id2sth("title", data_p)
god = "preprocess/godzilla.txt"
gods = ["m_"+l.strip() for l in open(god).readlines()]
for e_id in gods:
    print(e_id, id2title[e_id[2:]])
```

```python
num = 100
god_high_dim_embd = []
sizes = [0, 0, 0]

for e_id in gods:
    if e_id in bpt_item:
        god_high_dim_embd.append(bpt_item[e_id])
        sizes[0] += 1

for w in ['w_godzilla', "w_japan"]:
    god_high_dim_embd.append(bpt_word[w])
    sizes[1] += 1

shuffled =list(bpt_item.keys())
np.random.shuffle(shuffled)
for oth in shuffled:
    if num == 0: break
    if not oth in gods:
        god_high_dim_embd.append(bpt_item[oth])
        num -= 1
        sizes[2] += 1
    else:
        continue
        
god_2d = tsne.fit_transform(god_high_dim_embd)

pp = PdfPages('Godzilla_bpt_ty1font.pdf')
plt.figure(figsize=(10,8))
plt.clf()
plt.axis('off')
plt.margins(0.0)
zillas = plt.scatter(god_2d[:sizes[0],0], god_2d[:sizes[0],1], color="brown", s=80, marker='^', label="Godzilla movies", alpha=1)
z = plt.scatter(god_2d[sizes[0],0], god_2d[sizes[0],1], color="green", s=120, marker='+', label="Godzilla (word)", alpha=1)
japan = plt.scatter(god_2d[sizes[0]+1,0], god_2d[sizes[0]+1,1], color="green", s=80, marker='x', label="Japan (word)", alpha=1)
rand = plt.scatter(god_2d[-sizes[2]:,0], god_2d[-sizes[2]:,1], color="grey", s=30, marker='.', label="Other movies", alpha=0.5)
collection = [rand, zillas, japan, z]
plt.legend(handles=collection, loc=4, fontsize=20, scatterpoints=1)
pp.savefig(bbox_inches="tight")
pp.close()
plt.show()
```
